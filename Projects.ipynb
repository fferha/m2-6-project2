```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plot
import numpy as np
ins = pd.read_csv('C://Users/13439/Desktop/Python/new_project/Projects/1410002201_databaseLoadingData_ind2019.csv')
pt = ins[(ins.GEO != "Canada") & (ins.Sex != "Both sexes") & (ins.Age_group != "15 years and over")].pivot_table(index=['GEO'], columns='Age_group', values='VALUE', aggfunc='mean').round(2)
plot.rcParams['figure.figsize'] = (20, 8)
pt.plot(kind='bar',rot=45, title="Unemployment per Group of age due to Covid19",color = ['Pink', 'green', 'blue', 'Black'])
pt
```
